```r
rm(list=ls())
library(ggplot2)

data=read.table("combined_perf_metrics.txt",header=TRUE,sep='\t')
data
```

```r
options(repr.plot.width=10, repr.plot.height=8)
ggplot(data=data,
       aes(x=Metric,
           y=Value,
           fill=Substrate))+
  geom_boxplot()+
  scale_fill_manual(name="Substrate",values=c("#1b9e77","#d95f02","#7570b3"))+
  xlab("Performance Metric")+
  ylab("Peformance on\nHeld-Out Test Set")+
  ggtitle("n=20 random splits of train/test/validate sets")+
  theme_bw(20)+
  theme(axis.text.x = element_text(angle = 90))
```
